```python
import pandas as pd
# Source: http://web.mta.info/developers/turnstile.html
def get_data(mon):
    url = "citi/{}-citibike-tripdata.csv"
    dfs = []
    for mon in mon_nums:
        file_url = url.format(mon)
        dfs.append(pd.read_csv(file_url))
    return pd.concat(dfs)
        
mon_nums = [201812, 201901, 201903, 201904, 201905, 201906, 201907, 201908, 201909, 201910, 201911]
citi_df = get_data(mon_nums)
```

```python
citi_df.head()
```

```python
citi_df.drop("tripduration", axis=1, inplace=True)
```

```python
citi_df.drop("stoptime", axis=1, inplace=True)
citi_df.drop("starttime", axis=1, inplace=True)
citi_df.drop("bikeid", axis=1, inplace=True)
citi_df.drop("usertype", axis=1, inplace=True)
citi_df.drop("birth year", axis=1, inplace=True)
citi_df.drop("gender", axis=1, inplace=True)
citi_df.drop("counts", axis=1, inplace=True)
```

```python
citi_df['counts'] = 1
```

```python
# Sanity Check to verify that "C/A", "UNIT", "SCP", "STATION", "DATE_TIME" is unique
citi_df_sum = (citi_df
 .groupby(['start station id','start station name','start station latitude','start station longitude','end station id','end station name','end station latitude','end station longitude'])
 ['counts'].count()
 .reset_index()
 .sort_values("counts", ascending=False))
```

```python
citi_df_sum.to_csv('citi_df_sum.csv')
```
